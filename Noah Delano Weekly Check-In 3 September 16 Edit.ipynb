$$x_{n+1}=x_n-\frac{f(x_n)}{f'(x_n)}$$

Based on our function, we know:
$f(x)=x^2-a$ and
$f'(x)=2x$. With this in mind, we can sub this into our Newton Method:

$$x_{n+1}=x_n-\frac{f(x_n)}{f'(x_n)}$$

$$x_{n+1}=x_n-\frac{(x_n)^2-a}{2x_n}$$

$$x_{n+1}=\frac{2(x_n)^2-((x_n)^2-a)}{2x_n}$$

$$x_{n+1}=\frac{(x_n)^2+a}{2x_n}$$

$$x_{n+1}=\frac{1}{2}(x_n+\frac{a}{x_n}),\hspace{1cm} n\ge0$$

From that mindset, we can say:

$$Relative Error=\frac{True Value - Approximate Value}{True Value}$$

For our problem, we know that $x_{n+1}$ is our apporimate value and $\sqrt{a}$ is what our original value is. So:

$$Rel(x_{n+1})=\frac{\sqrt{a}-x_{n+1}}{\sqrt{a}}$$

And since we solved for $\sqrt{a}-x_{n+1}$ from the previous equation, we can sub it in and solve:

$$Rel(x_{n+1})=\frac{\sqrt{a}-x_{n+1}}{\sqrt{a}}$$

$$Rel(x_{n+1})=\frac{-\frac{1}{2x_n}(\sqrt{a}-x_n)^2}{\sqrt{a}}$$

But since the equation we are trying to derive has a factor of $Rel(x_n)$, then we need to get a factor of $\frac{\sqrt{a}-x_n}{\sqrt{a}}$. So:

$$Rel(x_{n+1})=-\frac{1}{2x_n}\frac{(\sqrt{a}-x_n)^2}{\sqrt{a}}$$

$$Rel(x_{n+1})=-\frac{1}{2x_n}\frac{(\sqrt{a}-x_n)^2}{\sqrt{a}}\frac{\sqrt{a}}{\sqrt{a}}$$

$$Rel(x_{n+1})=-\frac{\sqrt{a}}{2x_n}(\frac{\sqrt{a}-x_n}{\sqrt{a}})^2$$

$$Rel(x_{n+1})=-\frac{\sqrt{a}}{2x_n}[Rel(x_n)]^2$$

Now when we used the Newton Method, we would choose a starting value of the function, find the tangent line at the point, find the root of the tangent, and set that new x value as our approximation and rinse and repeat till the error is very little or non-existent. Halley's Method is similar, but uses hyperbola tangents rather than linear tangents.

So when we find the root of our tangent line, we approximate it at:

$$f(x_n)+f'(x_n)(x_{n+1}-x_n)=0$$

If you notice, this is a 1st degree taylor polynomial for the funtion f(x). So if we expand the approximation for a more accurate result, we can go to a 2nd degree taylor polynomial:

$$f(x_n)+f'(x_n)(x_{n+1}-x_n)+\frac{1}{2}f''(x_n)(x_{n+1}-x_n)^2=0$$

So now we have to solve for $x_{n+1}$ for our approximation:

$$f(x_n)+(x_{n+1}-x_n)[f'(x_n)+\frac{1}{2}f''(x_n)(x_{n+1}-x_n)]=0$$

$$(x_{n+1}-x_n)[f'(x_n)+\frac{1}{2}f''(x_n)(x_{n+1}-x_n)]=-f(x_n)$$

$$(x_{n+1}-x_n)=-\frac{f(x_n)}{f'(x_n)+\frac{1}{2}f''(x_n)(x_{n+1}-x_n)}$$

$$x_{n+1}=x_n-\frac{f(x_n)}{f'(x_n)+\frac{1}{2}f''(x_n)(x_{n+1}-x_n)}$$

However, we don't want the approximation we are solving for ($x_{n+1}$) in the equation. So to accomodate that, we can use Newton's Method (hence why Halley's Method is derived from Newton's) to sub in for $x_{n+1}$. So:

$$x_{n+1}=x_n-\frac{f(x_n)}{f'(x_n)+\frac{1}{2}f''(x_n)(x_n-\frac{f(x_n)}{f'(x_n)}-x_n)}$$

$$x_{n+1}=x_n-\frac{f(x_n)}{f'(x_n)-(\frac{f(x_n)f''(x_n)}{2f'(x_n)})}$$

$$x_{n+1}=x_n-\frac{f(x_n)}{(\frac{2(f'(x_n))^2-f(x_n)f''(x_n)}{2f'(x_n)})}$$

$$x_{n+1}=x_n-\frac{2f(x_n)f'(x_n)}{2[f'(x_n)]^2-f(x_n)f''(x_n)}$$

<span style = "color:red"> Nice </span>

So now we have derived Halley's Method. So lets do some programming to shoow it in action:

Lets use the function $f(x)=x^3-x^2-x-1$

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
##Define the function
def f(x):
    return x**3 - x**2 - x - 1
```

```python
#Define the derivative of the function
def df(x):
    return 3*x**2 - 2*x - 1
```

```python
#Define the second derivative of the function
def ddf(x):
    return 6*x - 2
```

```python
#Based on the graph, I want to start at x=4 for our intial guess

x0 = 4 # initial guess
ep = .00000005 # epsilon = allowed error

# Plot the function
#Create a list of x values
x = np.arange(-5,5,.01) 
#Apply the x values to our function
fplot = f(x) 
#Plot x and our f(x)
plt.plot(x,fplot) 
#Create a horizontal line at y=0 for finding roots
plt.hlines(0,-5,6) 
#Creates a vertical line at the x we are guessing
plt.vlines(x0,min(fplot),max(fplot),'r',linestyles='dashed')
plt.ylim(-10,10)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('f(x) and the initial guess x0')
plt.show()
```

```python
# Apply Halley's method
# Iterate until the error is less then the desired error
xold = x0
count = 0
err = 100 #Just a big number to help find an error less then ep
print('iteration  |   approx root  |   error ')
print('----------------------')

#Iterate till our error is below epsilon
while err > ep:
    #Apply the function and two derivates to the x we are guessing
    fxold = f(xold)
    dfxold = df(xold)
    ddfxold = ddf(xold)
    xnew = xold - ((2*fxold*dfxold)/(2*dfxold**2 - fxold*ddfxold)) # Halley's Method
    err = np.abs(xnew-xold) # Error
    
    if err > 100:
        print('The method is not converging!')
        break
    count += 1   #count = count + 1
    print(count, '  |  ', xnew, '  |  ', err)
    
    #Will plot the function, the tangent hyperbola, the guess for x, and their intersection
    x = np.arange(-5,5,.01)
    fplot = f(x)
    ftangent = fxold + dfxold*(x-xold) + (1/2)*ddfxold*(x-xold)**2
    plt.plot(x,fplot, label='f(x)')
    plt.plot(x, ftangent,'g-', label='Tangent at xold')
    plt.hlines(0,-3,6)
    plt.vlines(xold,-10,fxold,'r',linestyles='dashed', label='xold')
    plt.plot(xold,fxold,'mo', label='(xold,f(xold))')
    plt.xlim(-2,6)
    plt.ylim(-3,f(xold)+3)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('f(x) and the initial guess x0')
    plt.legend()
    plt.show()
    
    xold = xnew # update for the next iteration
    
print('----------------------')
print('Final Results:')
print(count, '  |  ', xnew, '  |  ', err)
```

```python
# Apply Newtons method to show differences and similarities
# Iterate until the error is less then the desired error
xold = x0
count = 0
err = 100 #Just a big number to help find an error less then ep
print('iteration  |   approx root  |   error ')
print('----------------------')

#Same Idea of previous cell, but for linear tangents instead
while err > ep:
    fxold = f(xold)
    dfxold = df(xold)
    xnew = xold - (fxold/dfxold) # Newton Method
    err = np.abs(xnew-xold) # Error
    
    if err > 100:
        print('The method is not converging!')
        break
    count += 1   #count = count + 1
    print(count, '  |  ', xnew, '  |  ', err)
    x = np.arange(-5,5,.01)
    fplot = f(x)
    ftangent = fxold + dfxold*(x-xold)
    plt.plot(x,fplot, label='f(x)')
    plt.plot(x, ftangent,'g-', label='Tangent at xold')
    plt.hlines(0,-3,6)
    plt.vlines(xold,-10,fxold,'r',linestyles='dashed', label='xold')
    plt.plot(xold,fxold,'mo', label='(xold,f(xold))')
    plt.xlim(-2,6)
    plt.ylim(-3,f(xold)+3)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('f(x) and the initial guess x0')
    plt.legend()
    plt.show()
    
    xold = xnew # update for the next iteration
    
    
print('----------------------')
print('Final Results:')
print(count, '  |  ', xnew, '  |  ', err)
```

As you can see, it works significantly faster than the Newton Method, with an epsilon significantly lower, Now this is only one example, but when dealing with functions with higher powers, creating hyperbola tangents definetly accelerate the amount of iterations needed to get a good approximation.I just wanted to share this method and compare it to Newton as I thought it was very interesting.
